```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('coins.jpg', cv2.IMREAD_GRAYSCALE)
_ , img = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(img, cmap= "gray")
```

```python
def top_crop(image):
    r,c = image.shape
    for i in range(r):
        for j in range(c):
            if image[i,j] !=255:
                return i,j

def bottom_crop(image):
    r,c = image.shape
    for i in range(r-1,0,-1):
        for j in range(c-1,0,-1):
            if image[i,j] !=255:
                return i,j

x1,y1 = top_crop(img)
x2,y2 = bottom_crop(img)

result = img[x1:x2 , y1:y2]
```

```python
contours, _ = cv2.findContours(result, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
```

```python
for c in contours:
    if len(c) > 100:
        x,y,w,h = cv2.boundingRect(c)
        cv2.rectangle(result, (x,y), (x+w,y+h), (127,127,127), 4)
plt.imshow(result, cmap='gray')        
```
