```python
import cv2
import matplotlib.pyplot as plt
```

```python
image1_original = cv2.imread('imgs/0.jpg')
image = cv2.imread('imgs/0.jpg', cv2.IMREAD_GRAYSCALE)
image = 255 - image
```

```python
plt.imshow(image , cmap='gray')
```

```python
_,img = cv2.threshold(image, 160, 255, cv2.THRESH_BINARY)
plt.imshow(image , cmap='gray')
```

```python
contours, _ = cv2.findContours(img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
k = 0
for c in contours :
    if len(c)>100:
        x,y,w,h = cv2.boundingRect(c)
        cv2.rectangle(img , (x,y), (x+w,y+h), (127,127,127), 4)
        k+=1
cv2.putText(image1_original, f"output: {k}", (50,60), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 2)       
plt.imshow(image1_original , cmap= "gray")
cv2.imwrite('output_imgs/result1.jpg',image1_original)
```

```python
image2_original = cv2.imread('imgs/1.jpg')
image2 = cv2.imread('imgs/1.jpg', cv2.IMREAD_GRAYSCALE)
image2 = 255 - image2
```

```python
plt.imshow(image2 , cmap='gray')
```

```python
_ , image2 = cv2.threshold(image2 , 10 , 255 , cv2.THRESH_BINARY)
plt.imshow(image2, cmap= 'gray')
```

```python
contours, _ = cv2.findContours(image2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

crop_rects = []
for c in contours :
    if len(c)>200 :
        x,y,w,h = cv2.boundingRect(c)
        rects = image2[y:y+h , x:x+w]
        crop_rects.append(255-rects)

plt.imshow(crop_rects[0], "gray")
plt.show()
plt.imshow(crop_rects[1], "gray")
plt.show()
```

```python
j=0
for rects in crop_rects:
    contours, _ = cv2.findContours(rects, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    n=0
    for c in contours:
        if 40< len(c) <70:
            x,y,w,h = cv2.boundingRect(c)
            cv2.rectangle(rects, (x,y), (x+w,y+h), (127,127,127), 4)
            n+=1
    cv2.putText(image2_original, f"output: {n}", (300-j,60), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 2)
    j+=200
plt.imshow(crop_rects[0], cmap= 'gray')
plt.show()
plt.imshow(crop_rects[1], cmap= 'gray')
plt.show()     
```

```python
cv2.imwrite('output_imgs/result2.jpg',image2_original)
plt.imshow(image2_original)
```

```python
image3_original = cv2.imread('imgs/2.jpg')
image3 = cv2.imread('imgs/2.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image3 , cmap= 'gray')
```

```python
image3 = 255-image3
plt.imshow(image3 , cmap='gray')
```

```python
_,img3 = cv2.threshold(image3 , 40 , 255 , cv2.THRESH_BINARY)
plt.imshow(img3 , cmap='gray')
```

```python
contours,_ = cv2.findContours(img3, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

crop_dics = []
for c in contours:
    x,y,w,h = cv2.boundingRect(c)
    dics = img3[y:y+h , x:x+w]
    crop_dics.append(255-dics)

#print(len(crop_dics))    

plt.imshow(crop_dics[0] , cmap= 'gray')
plt.show()
plt.imshow(crop_dics[1] , cmap= 'gray')
plt.show()
plt.imshow(crop_dics[2] , cmap= 'gray')
plt.show()
```

```python
j=0
for dic in crop_dics:
    contours,_ = cv2.findContours(dic, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    n=0
    for c in contours:
        if 110<len(c)<140:
            x,y,w,h = cv2.boundingRect(c)
            cv2.rectangle(dic, (x,y), (x+w,y+h), (127,127,127), 4)
            n+=1
    print(n)        
    cv2.putText(image3_original, f"output: {n}", (20,120-j), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 2)
    j+=30

plt.imshow(crop_dics[0] , cmap= 'gray')
plt.show()
plt.imshow(crop_dics[1] , cmap= 'gray')
plt.show()
plt.imshow(crop_dics[2] , cmap= 'gray')
plt.show()
```

```python
cv2.imwrite('output_imgs/result3.jpg',image3_original)
plt.imshow(image3_original)
```
