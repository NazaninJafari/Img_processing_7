```python
import cv2
import math
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('imgs/twitter.jpg' , cv2.IMREAD_GRAYSCALE)
_ , img = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(img, 'gray')
```

```python
rows, cols = img.shape
contours_list_j = []
contours_list_i = []

for i in range (rows):
    for j in range (cols):
        if img[i,j] == 255:
            contours_list_j.append(j)
            contours_list_i.append(i)
```

```python
j_min = min(contours_list_j) 
j_max = max(contours_list_j)

i_min = min(contours_list_i)
i_max = max(contours_list_i)

y,x = i_min,j_min
h = i_max - i_min
w = j_max - j_min

cv2.rectangle(img, (x,y) , (x+w , y+h), (127,127,127), 9)
```

```python
plt.imshow(img, cmap = "gray")
cv2.imwrite('output_imgs/twitter_result.jpg', img)
```
